```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df = pd.read_excel("/content/Data_Train.xlsx")
train_df.head()
```

```python
test_df = pd.read_excel("/content/Test_set.xlsx")
test_df.head()
```

```python
# merging train and test df
df = train_df.append(test_df)
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df['Date'] = df["Date_of_Journey"].apply(lambda x:x.split('/')[0])
df['Month'] = df["Date_of_Journey"].apply(lambda x:x.split('/')[1])
df['Year'] = df["Date_of_Journey"].apply(lambda x:x.split('/')[2])
```

```python
df.head()
```

```python
df.info()
```

```python
# now converting date , month , and year in int.
```

```python
df['Date'] = df["Date"].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df["Year"].astype(int)
```

```python
df.info()
```

```python
# dropping the Date of Journey column
df.drop("Date_of_Journey",axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x : x.split(" ")[0])
# removed month from arrival_time
```

```python
df.head(2)
```

```python
df['Arrival_Hour'] = df["Arrival_Time"].apply(lambda x : x.split(':')[0])
df['Arrival_Min'] = df["Arrival_Time"].apply(lambda x : x.split(':')[1])
```

```python
df.head(2)
```

```python
df.info()
```

```python
df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df["Arrival_Min"] = df['Arrival_Min'].astype(int)
```

```python
df.info()
```

```python
df['Dep_Hour'] = df['Dep_Time'].apply(lambda x : x.split(':')[0])
df['Dep_Min'] = df["Dep_Time"].apply(lambda x : x.split(':')[1])
```

```python
df.head(2)
```

```python
df['Dep_Hour'] = df['Dep_Hour'].astype(int)
df['Dep_Min'] = df["Dep_Min"].astype(int)
```

```python
df.drop("Dep_Time",axis=1,inplace=True)
```

```python
df.info()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```
